```python
# Explore the data
df.head()
```

### Explore the Data
It looks like the data was read in correctly, but notice that the first 3 rows are not data - but metadata.

The first row that pandas read in is the column names - so only skip the first 2 rows.

`read_excel` has a `skiprows` option that can be used to skip the first 2 rows.

```python
# skip the first 2 rows in read excel file
df = pd.read_excel("..\SourceData\Atmos41WeatherStation\ATMOS41_2022-12-23.xlsx", skiprows=2)
```

```python
# Explore the data
df.head()
```

```python
# Explore the data - describe all variables
df.describe().T
```

## Read in a Large CSV file

Help from Google Searches:
- https://medium.com/analytics-vidhya/optimized-ways-to-read-large-csvs-in-python-ab2b36a7914e 

Dask documentation:
- https://docs.dask.org/en/latest/dataframe.html
- help with Dask: https://www.youtube.com/watch?v=jstCmSD_LAs

```python
from dask import dataframe as dd

# Read in data and set data type to object (which is tha same as a string)
# The csv file can have mixed data types and this will allow us to read in the data with fewer errors 
dask_df = dd.read_csv('..\SourceData\OpenFEMA\IndividualsAndHouseholdsProgramValidRegistrations (1).csv', 
                    low_memory=False,
                    dtype= 'object')
```

```python
dask_df
```

```python
dask_df.head(3)
```

```python
# locate by disaster number and county
condition1 = (dask_df['disasterNumber'] == '1791')
condition2 = (dask_df['county'] == 'Galveston (County)')
drnum_county_ddf = dask_df.loc[condition1 & condition2].compute()
```

```python
drnum_county_ddf.head()
```

```python
drnum_county_ddf.groupby('county').size()
```
