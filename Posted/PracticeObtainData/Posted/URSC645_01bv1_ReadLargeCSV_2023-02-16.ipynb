### Make a function to clean variables
Create a function to clean variables. This function can be used with any dataset and set of variables

```python
def clean_variables(df, var_list, var_type):
    '''
    This function will convert a list of 
    variables to a specified data type

    '''

    # Check variables 
    for var in var_list:
        print('Variable: ' + str(var))
        # check if variable is a string
        if df[var].dtype == object:
            print('Variable is an object (string)')
            # Check categories
            # Group by variable to see the number of unique values
            print(df.groupby(var).size())
            # check if desired data type is an integer or float
            if var_type == int or var_type == float:
                try:
                    print('Converting to variable type: ' + str(var_type))
                    # convert to specified data type
                    df[var] = df[var].astype(var_type)
                except:
                    print('Could not convert to variable type: ' + str(var_type))
            else:
                print('Converting to variable type: ' + str(var_type))
                # convert to specified data type
                df[var] = df[var].astype(var_type)
        else:
            print('Converting to variable type: ' + str(var_type))
            # convert to specified data type
            df[var] = df[var].astype(var_type)
    return df
```

```python
var_list = ['occupantsUnderTwo','occupants2to5','occupants6to18','occupants19to64']
var_type = int
drnum_county_df = clean_variables(drnum_county_df, var_list, var_type)
```
